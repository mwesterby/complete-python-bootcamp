# Working with CSV Files

### Reading from a CSV

```python
import csv
```

```python
# Open the file
data = open('example.csv', encoding='utf-8')
```

```python
# csv.reader
csv_data = csv.reader(data)
```

```python
# reformat it into a python object list of lists
data_lines = list(csv_data)
```

```python
data_lines
```

```python
data_lines[0]
```

Column names

```python
len(data_lines)
```

1000 rows (+ 1 row for header)

```python
for line in data_lines:
    print(line)
```

```python
data_lines[10][3] # Row 10, third cell
```

```python
all_emails = []
for line in data_lines[1:15]: # Start at 1 because row 0 is column headings
    all_emails.append(line[3])
```

```python
all_emails
```

```python
data_lines[10]
```

```python
full_names = []
for line in data_lines[1:]:
    full_names.append(line[1] + " " + line[2])
```

```python
full_names
```

### Writing to a CSV

```python
file_to_output = open('to_save_file.csv', mode='w', newline='')
```

Mode could be append ` mode='a'` to append to the file

```python
csv_writer = csv.writer(file_to_output, delimiter=',')
```

```python
csv_writer.writerow(['a', 'b', 'c'])
```

```python
csv_writer.writerows([['1', '2', '3'], ['4', '5', '6']])
```

```python
file_to_output.close()
```

```python
f = open('to_save_file.csv', mode='a', newline='')
```

```python
csv_writer = csv.writer(f)
```

```python
csv_writer.writerow(['1', '2', 3])
```

Output here is the number of characters written

```python
f.close()
```
